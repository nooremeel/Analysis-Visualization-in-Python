```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
movie_data=pd.read_csv('movie.csv')
movie_data
```

```python
movie_df = pd.DataFrame(movie_data)
print(movie_df.head())
```

```python
movie_df.info()
```

```python
# Check for columns with null values
null_columns = movie_df.columns[movie_df.isnull().any()]
null_columns
```

```python
#replace values only in gross with mean value
gross_mean = movie_df['gross'].mean()
movie_df['gross'] = movie_df['gross'].fillna(gross_mean)
movie_df
```

```python
#replace empty numeric values with 0 and empty string values with 'null'
def replace_null(val):
    if pd.isnull(val) or val == '':
        if isinstance(val, str) or val == '':
            return 'null'
        else:
            return 0
    return val

movie_df = movie_df.applymap(replace_null)
```

```python
movie_df['director_name'] = movie_df['director_name'].replace(0, 'unknown')
```

```python
movie_df = movie_df[movie_df['duration'] >= 0]
negative_durations = movie_df[movie_df['duration'] < 0]

print(negative_durations)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(movie_df['duration'], movie_df['imdb_score'], marker='o', linestyle='-', color='r', s=10)
plt.xlabel('duration')
plt.ylabel('rating')
plt.title('relation between duration and rating')
plt.show()
```

```python
aggregated_data = movie_df.groupby('duration')['imdb_score'].mean().reset_index()

plt.figure(figsize=(12, 6))

plt.plot(aggregated_data['duration'], aggregated_data['imdb_score'], marker='o')

plt.xlabel('duration')
plt.ylabel('average rating')
plt.title('average rating by movie duration')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.scatter(movie_df['budget'], movie_df['gross'], marker='o', linestyle='-', color='r', s=10)
plt.xlabel('budget')
plt.ylabel('gross')
plt.title('relation between budget and gross')
plt.show()
```

```python
aggregated_data = movie_df.groupby('budget')['gross'].mean().reset_index()
plt.figure(figsize=(12, 6))

plt.plot(aggregated_data['budget'], aggregated_data['gross'], marker='o')

plt.xlabel('')
plt.ylabel('average aating')
plt.title('relation between budget and gross')
plt.show()
```

```python
aggregated_data = movie_df.groupby('duration')['imdb_score'].mean().reset_index()


plt.figure(figsize=(12, 6))
sns.lineplot(x='duration', y='imdb_score', data=aggregated_data, marker='o')

plt.xlabel('duration')
plt.ylabel('average rating')
plt.title('Average Rating by Movie Duration')

plt.show()
```

```python
duration_bins = pd.cut(movie_df['duration'], bins=20)
rating_bins = pd.cut(movie_df['imdb_score'], bins=10)

heatmap_data = pd.pivot_table(movie_df, index=rating_bins, columns=duration_bins, aggfunc='size', fill_value=0)


plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d')

plt.xlabel('Duration (binned)')
plt.ylabel('Rating (binned)')
plt.title('Heatmap: Movie Count by Duration and Rating')

plt.show()
```

```python
director_counts = movie_df[movie_df['director_name']!= 'unknown']['director_name'].value_counts()

top_directors = director_counts.head(10)
plt.figure(figsize=(10, 6))
top_directors.plot(kind='bar')

plt.xlabel('director')
plt.ylabel('number of movies')
plt.title('Top 10 Most Frequent Directors')

plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
top_movies = movie_df.sort_values(by='gross', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top_movies['movie_title'], top_movies['gross'], color='green')

plt.xlabel('movie name')
plt.ylabel('gross')
plt.title('Top 10 Most Grossing Movies')

plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
average_rating_per_year = movie_df.groupby('title_year')['imdb_score'].mean().reset_index()
average_rating_per_year.columns = ['title_year', 'Average Rating']
top_years = average_rating_per_year.sort_values(by='Average Rating', ascending=False).head(25)

plt.figure(figsize=(14, 6))
plt.bar(top_years['title_year'].astype(str), top_years['Average Rating'], color='skyblue')

plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Average Rating per Year (top 25)')
plt.xticks(rotation=45)

plt.show()
```
